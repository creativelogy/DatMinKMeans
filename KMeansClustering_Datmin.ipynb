```python
# import libraries
import pandas as pd
import math as mt

# membaca data csv
data = pd.read_csv('https://raw.githubusercontent.com/creativelogy/DatMinKMeans/main/DataSetDatmin.csv')

# rumus euclidian distance
def euclidian_distance(klaster, data, idxLama):
  cols = ['k1', 'k2', 'k3', 'k4']
  df = pd.DataFrame(columns=cols, index=range(data.shape[0]))
  for i in range(data.shape[0]):
    hasil = []
    for j in range(klaster.shape[0]):
      hasil.append(mt.sqrt(((klaster['fitur_x'].iloc[j] - data['fitur_x'].iloc[i])**2)+
                          ((klaster['fitur_y'].iloc[j] - data['fitur_y'].iloc[i])**2)+
                          ((klaster['fitur_z'].iloc[j] - data['fitur_z'].iloc[i])**2)))
    df.loc[i] = hasil
  print()
  print(df)
  return jarak_terdekat(data, df, idxLama)

# menentukan jarak terdekat ke klaster
def jarak_terdekat(data, df, idxLama):
  cols = ['k1', 'k2', 'k3', 'k4']
  terdekat = pd.DataFrame(columns=cols, index=range(data.shape[0]))
  for i in range(data.shape[0]):
    smallest = min(df['k1'].iloc[i],df['k2'].iloc[i],df['k3'].iloc[i],df['k4'].iloc[i])
    if (smallest == df['k1'].iloc[i]):
      terdekat.loc[i] = [1, 0, 0, 0]
    elif (smallest == df['k2'].iloc[i]):
      terdekat.loc[i] = [0, 1, 0, 0]
    elif (smallest == df['k3'].iloc[i]):
      terdekat.loc[i] = [0, 0, 1, 0]
    elif (smallest == df['k4'].iloc[i]):
      terdekat.loc[i] = [0, 0, 0, 1]
  print()
  print(terdekat)
  return index_terdekat(data, terdekat, idxLama)

# memasukkan index mana saja yang terdekat dengan klaster ke dalam variabel
def index_terdekat(data, terdekat, idxLama):
  idxBaru_k1 = []
  idxBaru_k2 = []
  idxBaru_k3 = []
  idxBaru_k4 = []
  for i in range(data.shape[0]):
    if (terdekat.loc[i].k1 == 1):
      idxBaru_k1.append(i)
    elif (terdekat.loc[i].k2 == 1):
      idxBaru_k2.append(i)
    elif (terdekat.loc[i].k3 == 1):
      idxBaru_k3.append(i)
    elif (terdekat.loc[i].k4 == 1):
      idxBaru_k4.append(i)
  idxBaru = [idxBaru_k1, idxBaru_k2, idxBaru_k3, idxBaru_k4] 
  print()
  print("index baru: ", idxBaru)
  print("index lama: ", idxLama)
  print()
  return cek_index(idxLama, idxBaru, data, idxBaru_k1, idxBaru_k2, idxBaru_k3, idxBaru_k4)

# cek idx lama apakah sama dengan idx baru
def cek_index(idxLama, idxBaru, data, idxBaru_k1, idxBaru_k2, idxBaru_k3, idxBaru_k4):
  if (idxLama == idxBaru):
    print("Yey Kelar")
    print("Klaster 1:", idxBaru_k1)
    print("Klaster 2:", idxBaru_k2)
    print("Klaster 3:", idxBaru_k3)
    print("Klaster 4:", idxBaru_k4)
  else:
    print("index lama != index baru")
    idxLama = idxBaru
    return klaster_baru(idxLama, data, idxBaru_k1, idxBaru_k2, idxBaru_k3, idxBaru_k4)


# fungsi untuk declare klaster baru
def klaster_baru(idxLama, data, k1, k2, k3, k4):
  result_x1 = result_y1 = result_z1 = result_x2 = result_y2 = result_z2 = 0
  result_x3 = result_y3 = result_z3 = result_x4 = result_y4 = result_z4 = 0
  for i in k1:
    result_x1 = result_x1 + data['fitur_x'].iloc[i]/len(k1)
    result_y1 = result_y1 + data['fitur_y'].iloc[i]/len(k1)
    result_z1 = result_z1 + data['fitur_z'].iloc[i]/len(k1)
  for i in k2:
    result_x2 = result_x2 + data['fitur_x'].iloc[i]/len(k2)
    result_y2 = result_y2 + data['fitur_y'].iloc[i]/len(k2)
    result_z2 = result_z2 + data['fitur_z'].iloc[i]/len(k2)
  for i in k3:
    result_x3 = result_x3 + data['fitur_x'].iloc[i]/len(k3)
    result_y3 = result_y3 + data['fitur_y'].iloc[i]/len(k3)
    result_z3 = result_z3 + data['fitur_z'].iloc[i]/len(k3)
  for i in k4:
    result_x4 = result_x4 + data['fitur_x'].iloc[i]/len(k4)
    result_y4 = result_y4 + data['fitur_y'].iloc[i]/len(k4)
    result_z4 = result_z4 + data['fitur_z'].iloc[i]/len(k4)

  result_x = [result_x1, result_x2, result_x3, result_x4]
  result_y = [result_y1, result_y2, result_y3, result_y4]
  result_z = [result_z1, result_z2, result_z3, result_z4]

  klasterBaru = pd.DataFrame(
      {'fitur_x':result_x, 'fitur_y':result_y, 'fitur_z':result_z},
      index = ['k1', 'k2', 'k3', 'k4']
  )
  print()
  print("klaster baru")
  print(klasterBaru)
  return euclidian_distance(klasterBaru, data, idxLama) 

# print data awal
print(data)
print()

# declare klaster awal
data_klaster = pd.DataFrame(
    {'fitur_x':[data['fitur_x'].iloc[3],
                data['fitur_x'].iloc[7], 
                data['fitur_x'].iloc[11], 
                data['fitur_x'].iloc[15]],
      'fitur_y':[data['fitur_y'].iloc[3], 
                data['fitur_y'].iloc[7], 
                data['fitur_y'].iloc[11], 
                data['fitur_y'].iloc[15]],
      'fitur_z':[data['fitur_z'].iloc[3], 
                data['fitur_z'].iloc[7], 
                data['fitur_z'].iloc[11], 
                data['fitur_z'].iloc[15]],
      }, index=['k1', 'k2', 'k3', 'k4'])

print(data_klaster)

# declare index lama untuk memasukkan hasil index terdekat guna membandingkan
# indeks terdekat selanjutnya, dan memanggil rumus euclidian distance
idxLama = []
euclidian_distance(data_klaster, data, idxLama)
```
