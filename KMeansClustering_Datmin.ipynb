```python
# DATA MINING - Khalifah Imam Ali - M0519054 - Kelas B

from random import randrange
import pandas as pd
import math as mt


Data = tuple[int, int, int]
Clusters = dict[Data, list[Data]]

# Dataset terdiri dari 3 row yang mana mewakili -> fitur x, fitur y, fitur z
dataset = pd.read_csv('https://raw.githubusercontent.com/creativelogy/DatMinKMeans/main/DataSetDatmin.csv')


# Pertama kita melakukan grouping
def group_data_into_clusters(cluster_centers: list[Data]) -> Clusters:
    cluster_groups = {}
    for center in cluster_centers:
        cluster_groups[center] = []

    for data in dataset:
        distances = []
        for center in cluster_centers:
            quadratic_sum = (
                
                (center[0] - data[0]) ** 2
                + (center[1] - data[1]) ** 2
                + (center[2] - data[2]) ** 2

            )
            distance = quadratic_sum ** 0.5
            distances.append((center, distance))

        #Tentukan center yang terdekat
        distances.sort(key=(lambda d: d[1]))
        shortest_center = distances[0][0]

        cluster_groups[shortest_center].append(data)
    #Kembalikan nilai grup klaster
    return cluster_groups


# menghitung mean (rata-rata) untuk tiap pusat klaster
def calculate_new_cluster_centers(clusters: Clusters) -> list[Data]:
    new_cluster_centers = []
    
    for cluster_data in clusters.values():
        new_center = [0, 0, 0]

        for data in cluster_data:
            new_center[0] += data[0]
            new_center[1] += data[1]
            new_center[2] += data[2]

        for i, val in enumerate(new_center):
            new_center[i] = val / len(cluster_data)
        
        #ganti cluster center baru
        new_cluster_centers.append(tuple(new_center))

    return new_cluster_centers

# menentukan cluster center pertama kali
def get_random_cluster_centers(n=1) -> list[Data]:
    cluster_centers = []
    while len(cluster_centers) != n:
        i = randrange(len(dataset))
        if i not in cluster_centers:
            cluster_centers.append(dataset[i])
    
    return cluster_centers


# menunjukan hasil center klaster 
def display_cluster_centers(cluster_centers: list[Data]):
    print(f'Pusat cluster:\n{cluster_centers}\n')

#Fungsi utama
def main():
    #Misalnya jumlah klastering kita adalah sebanyak 4
    NUM_CLUSTER = 4
    RANDOM_INIT = False

    #Definisikan dulu bahwa cluster centernya tidak ada di awal
    cluster_centers = []

    #Lakukan randomize pusat klaster dengan fungsi dibawah
    if RANDOM_INIT:
        cluster_centers = get_random_cluster_centers(NUM_CLUSTER)
    else:
        cluster_centers = [dataset[i + 10] for i in range(4)]

    print('      START      ')
    print('===============\n')

    while True:
        #Menunjukan pusat klaster
        display_cluster_centers(cluster_centers)

        #Melakukan gruping data dan menentukan klaster baru setelah dihitung
        clusters = group_data_into_clusters(cluster_centers)
        new_cluster_centers = calculate_new_cluster_centers(clusters)

        #Setelah melakukan looping, maka akan dimuncuklan pusat klaster akhir
        #ketika pusat klaster tidak lagi berubah-ubah
        if cluster_centers == new_cluster_centers:
            print('================')
            print('  FINAL RESULT  ')
            print('================\n')
            
            #Menampilkan cluster akhir
            for i, (center, cluster_data) in enumerate(clusters.items()):
                print(f'{i + 1}. Cluster {i + 1}, pusat di {center}:\n')
                print('\t-> ', end='')
                for data in cluster_data:
                    print(data, end=', ')
                print('\n')
            break
        else:
            cluster_centers = new_cluster_centers
            
if __name__ == '__main__':
    main()
```
